```python
import os
import glob
from typing import Dict, List, Tuple, Optional

import numpy as np
import xarray as xr

import matplotlib.pyplot as plt

from pcmdi_enso_reader import (
    ENSODiagReader
)
```

```python
class ENSOAmplitudePlotter:
    """
    Plot ENSO amplitude diagnostics (e.g., SSTA std vs longitude)
    from ds_amp produced by ENSODiagReader.load_hist_future and
    an observational reference curve.
    """

    def __init__(
        self,
        ds_amp: xr.Dataset,
        fig_dir: str,
        obs_da: Optional[xr.DataArray] = None,
        obs_label: str = "ref: ERA-Interim",
        model_labels: Optional[dict] = None,
    ):
        self.ds_amp = ds_amp
        self.fig_dir = fig_dir
        self.obs_da = obs_da
        self.obs_label = obs_label
        
        if model_labels is None:
            self.model_labels = {
                str(p): str(p) for p in self.ds_amp["period"].values
            }
        else:
            self.model_labels = model_labels

    def _prep_amp(self) -> xr.DataArray:
        """
        Return amplitude DataArray from ds_amp['metric'].
        Dims: (period, member, longitude)
        """
        da = self.ds_amp["metric"]
        extra_dims = [d for d in da.dims if d not in ("period", "member", "longitude")]
        if extra_dims:
            da = da.mean(dim=extra_dims)
        return da
        
    @staticmethod
    def _format_lon(lon):
        """
        Convert longitude in degrees to string label (e.g., -80 → 80°W, 80 → 80°E).
        Assumes lon is in [-180, 180].
        """
        lon = float(lon)
        if lon == 0 or lon == 360:
            return "0°"
        if lon == 180:
            return "180°"
        if lon < 180:
            return f"{int(lon)}°E"
        else:
            return f"{int(360 - lon)}°W"
            
    def plot_var_lon(
        self,
        filename: str = "enso_amplitude_hist_future_ens.pdf",
        ymin: Optional[float] = None,
        ymax: Optional[float] = None,
        show_spread: bool = True,
        title: str = "ENSO amplitude (SSTA standard deviation)",
        xlabel: str = "longitude",
        ylabel: str = r"SSTA std ($^\circ$C)",
        xint: int = 30, 
        figsize: tuple = (5, 4),
        fontz: int = 12,
    ):
        da = self._prep_amp()               # (period, member, longitude)
        lon = da["longitude"].values
    
        fig, ax = plt.subplots(figsize=figsize)
    
        # --- font sizes ---
        ax.set_xlabel(xlabel, fontsize=fontz * 1.0)
        ax.set_ylabel(ylabel, fontsize=fontz * 1.0)
        ax.set_title(title,  fontsize=fontz * 1.1, loc = "left")
        ax.tick_params(axis="both", labelsize=fontz * 0.95)
    
        # --- observations ---
        if self.obs_da is not None:
            obs = self.obs_da
            if "longitude" in obs.dims:
                obs_vals = obs.sel(longitude=lon, method="nearest")
            else:
                obs_vals = obs
            ax.plot(lon, obs_vals, linewidth=2, color="black", label=self.obs_label)
    
        # --- hist & future ensemble curves ---
        for pval in da["period"].values:
            label = self.model_labels.get(str(pval), str(pval))
            da_p = da.sel(period=pval)          # (member, longitude)
            mean_p = da_p.mean("member")
            std_p  = da_p.std("member")
    
            line = ax.plot(lon, mean_p, linewidth=2, label=f"{label}")[0]
            color = line.get_color()
    
            if show_spread:
                ax.fill_between(
                    lon, mean_p - std_p, mean_p + std_p,
                    color=color, alpha=0.2, linewidth=0,
                )
    
        # --- longitude ticks / labels, assuming [-180, 180] ---
        lon_min = float(lon.min())
        lon_max = float(lon.max())
        lon_min_rounded = np.floor(lon_min / xint) * xint
        lon_max_rounded = np.ceil(lon_max / xint) * xint
        xticks = np.arange(lon_min_rounded, lon_max_rounded + 1e-6, xint)
    
        ax.set_xticks(xticks)
        ax.set_xticklabels(
            [self._format_lon(x) for x in xticks],
            fontsize=fontz * 0.95,
        )
    
        if ymin is not None or ymax is not None:
            ax.set_ylim(ymin, ymax)
        ax.grid(True, linestyle="--", alpha=0.5)
        ax.legend(loc="best", frameon=True, fontsize=fontz * 0.9)
    
        out_path = os.path.join(self.fig_dir, filename)
        fig.tight_layout()
        fig.savefig(out_path, dpi=150, format="pdf")
        print(f"Saved ENSO amplitude diagnostic figure to {out_path}")
        return fig, ax
```

```python
if __name__ == "__main__":
    TOP_DIR   = "/lcrc/group/e3sm/ac.szhang/acme_scratch/e3sm_project/v3_le_paper"
    DATA_DIR  = "/lcrc/group/e3sm/public_html/diagnostic_output/ac.szhang/e3sm-pcmdi-le"
    OUT_DIR   = f"{TOP_DIR}/figure_data"
    FIG_DIR   = "./"
    os.makedirs(OUT_DIR, exist_ok=True)
    os.makedirs(FIG_DIR, exist_ok=True)

    # ---- Pair experiments consistently (order matters) ----
    MODEL   = "v3.LR.historical"
    GROUP   = ["hist", "future"]
    PERIOD  = [(1985, 2014), (2015, 2049)]
    NENS    = [25, 25]

    members = None
    verbose = False
    diag_print = True 
    
    reader = ENSODiagReader(
        data_dir=DATA_DIR,
        model=MODEL,
        groups=GROUP,
        period_list=PERIOD,
        nens=NENS,
        members=members,
        verbose=verbose,
    )

    target_obs    = "ERA-Interim"
    target_group  = "ENSO_perf"
    target_metric = "enso_amplitude"
    target_var_list = [
        "sstStd_lon" 
    ]

    # ---- Sanity checks ----
    print("Available ENSO groups:", reader.available_groups())
    print(f"Variables in {target_group}:", reader.available_vars(target_group))

    if target_group not in reader.available_groups():
        raise ValueError(f"Unknown target_group: {target_group}")

    if target_metric not in reader.available_vars(target_group):
        raise ValueError(f"Unknown target_metric '{target_metric}' for '{target_group}'")

    data_dict = {}   # model hist+ model future + obs 
    # ----------------------------------------------------
    # Loop through target variable list
    # ----------------------------------------------------
    for target_var in target_var_list:
        print(f"\n>>> Loading target_var = {target_var}")
        data_dict[target_var] = {} 
        
        # Load model + obs using your working function
        dm, do = reader.load_metric_data(
            enso_group=target_group,
            var_name=target_metric,
            nc_var=target_var,
            ref_tag=target_obs,
        )

        # observation only has one member and is identical for historical and near-future
        ref = reader.validate_constant_observation(
            do, ref_group="hist", ref_member="00",  sample_dim='longitude'
        )
        # If we get here, all obs are identical
        # ref is the DataArray you want to keep
        data_dict[target_var]['reference'] = ref
        
        for group, member_dict in dm.items():
            print(f"  Group: {group}")
            pooled = reader.pool_members_to_samples(member_dict, sample_dim='longitude')  # sample_dim inferred
            data_dict[target_var][group] = pooled
            
    print("\n✓ All target_var successfully read.")

    if diag_print: 
        for var, group_dict in data_dict.items():
            print(f"\nVariable: {var}")
            for group, da in group_dict.items():
                print(f"  Group: {group}")
                # da is an xarray DataArray or Dataset
                shape = getattr(da, "shape", None)
                dims  = getattr(da, "dims", None)
                print(f"type={type(da).__name__} | shape={shape} | dims={dims}")
```

```python
if __name__ == "__main__":
    TOP_DIR   = "/lcrc/group/e3sm/ac.szhang/acme_scratch/e3sm_project/v3_le_paper"
    DATA_DIR  = "/lcrc/group/e3sm/public_html/diagnostic_output/ac.szhang/e3sm-pcmdi-le"
    OUT_DIR   = f"{TOP_DIR}/figure_data"
    FIG_DIR   = "./"
    os.makedirs(OUT_DIR, exist_ok=True)
    os.makedirs(FIG_DIR, exist_ok=True)

    # ---- Pair experiments consistently (order matters) ----
    MODEL   = "v3.LR.historical"
    GROUP   = ["hist", "future"]
    PERIOD  = [(1985, 2014), (2015, 2049)]
    NENS    = [25, 25]
    
    reader = ENSODiagReader(
        data_dir=DATA_DIR,
        model=MODEL,
        groups=GROUP,
        period_list=PERIOD,
        nens=NENS,
        # enso_groups and file_suffix_map default to internal getters
        members=None,   # or [51, 91, 101, ...] if you want explicit list
        verbose=False,  # set True if you want to see directory paths
    )

    target_obs    = "ERA-Interim"
    target_group  =  "ENSO_perf" #"ENSO_proc" # "ENSO_tel"
    target_metric = "enso_amplitude"   # logical metric key (controls which file)
    target_var    = "sstStd_lon"       # base pattern inside the file

    # Optional: quick sanity checks using helper methods
    print("Available ENSO groups:", reader.available_groups())
    print(f"Variables in {target_group}:", reader.available_vars(target_group))

    if target_group not in reader.available_groups():
        raise ValueError(f"Unknown target_group: {target_group}")

    # check the metric key, not the base var pattern
    if target_metric not in reader.available_vars(target_group):
        raise ValueError(
            f"Unknown target_metric '{target_metric}' for group '{target_group}'"
        )

    # Load model ENSO amplitude vs longitude for hist+future
    # nc_var is treated as a BASE PATTERN (e.g. "sstStd_lon")
    dm, do = reader.load_metric_data(
        enso_group=target_group,
        var_name=target_metric,
        nc_var=target_var,
        ref_tag=target_obs,
    )
    print(dm)
    print(do)
    
    # Note that do here is the same for all members in the future and historical periods
    # Compare every entry to do[0,0]
    ref = do.isel(period=0, member=0)
    # Boolean mask: True if identical, False otherwise
    all_same = bool((do == ref).all())
    if not all_same:
        raise ValueError("Observation dataset `do` differs across periods or members. "
                         "Your assumption in the note is NOT valid.")
    # If all identical:
    obs_da = do["metric"].isel(period=0, member=0)   # dims: (longitude) or (..., longitude)
    print("Observation dataset is identical across all members and periods. Using the first one.")
        
    # 3. Instantiate plotter
    model_labels = {        
        "hist": "E3SM (Historical)",
        "future": "E3SM (Near-Future)",
    }
    
    obs_label=f"{target_obs} (reference)"
    
    figure_name = "enso_amplitude_hist_future_ens.pdf"
    ymin=0.0
    ymax=2.0
    show_spread=True
    title = "(a) ENSO Amplitude (SSTA standard deviation)"
    xlabel = "Longitude"
    ylabel = r"SSTA standard deviation($^\circ$C)"
    fontz  = 18 
    figure_size = (10,8)
    
    plotter = ENSOAmplitudePlotter(
        ds_amp=dm,
        fig_dir=FIG_DIR,
        obs_da=obs_da,
        obs_label=obs_label,
        model_labels=model_labels,
    )

    # 4. Make the figure
    plotter.plot_var_lon(
        filename=figure_name,
        ymin=ymin,
        ymax=ymax,
        title=title,
        xlabel=xlabel,
        ylabel=ylabel,
        fontz=fontz,
        figsize=figure_size,
        show_spread=show_spread,
    )
```
